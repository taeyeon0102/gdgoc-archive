```python
## 2개의 층
# 1. 데이터 가져오기
from tensorflow import keras
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()
```

```python
# 2. input data 준비
from sklearn.model_selection import train_test_split
train_scaled = train_input/ 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42)
```

```python
# 3. layer 생성
# 입력(784개 특성) -> 100개 뉴런이 있는 layer(100개의 특징으로 압축) -> 10개 분류 layer -> 출력층
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,))
dense2 = keras.layers.Dense(10, activation='softmax')
```

```python
# 4. 심층 신경망(DNN)
model = keras.Sequential([dense1, dense2])
model.summary() #정보 출력
#모델 이름/각 layer의 이름, 클래스, 출력 크기, 모델 파라미터 개수
```

```python
# 5. 층을 추가하는 다른 방법
# Sequential 클래스의 생성자 안에서 바로 Dense 클래스의 객체를 만듦(보통)
model = keras.Sequential([
    keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'),
    keras.layers.Dense(10, activation='softmax', name='output')
], name='Fashion MNIST Model')
model.summary() #정보 출력
```

```python
# 6. add()로 layer 추가
model = keras.Sequential()
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary() #정보 출력
```

```python
# 7. model train
print(train_scaled.shape) #input data 확인
train_scaled = train_scaled.reshape(-1, 28*28)
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# 8. Flatten layer
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()
```

```python
# 9. Flatten Layer을 포함한 모델
# 훈련 세트와 테스트 세트 나누기 but reshape() 메서드 적용x
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()
train_scaled = train_input/ 255.0
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_input, train_target, test_size=0.2, random_state=42)
# 모델 컴파일하고 훈련
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# 검증 세트로 성능 확인
model.evaluate(val_scaled, val_target)
```

```python
## Adam optimizer을 이용하여 모델 훈련
# 1. 모델  생성
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

# 2. 훈련 (optimizer = Adam, ephocs = 5)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
# 3. 평가
model.evaluate(val_scaled, val_target)
```
